- High popularity songs
- Low popularity songs

In this notebook, the datasets are manually uploaded using Google Colab's file upload functionality.

No additional data collection or scraping was required.

After loading the datasets, basic checks such as dataset shape, column names, and missing values were performed to ensure data quality before proceeding to analysis.

```python
import pandas as pd

high = pd.read_csv("high_popularity_spotify_data.csv")
low = pd.read_csv("low_popularity_spotify_data.csv")

print("High popularity songs:", high.shape)
print("Low popularity songs:", low.shape)

print("\nColumns:\n")

for i, col in enumerate(high.columns):
    print(i, "-", col)

print("\nMissing values:\n")
print(high.isnull().sum())
```
